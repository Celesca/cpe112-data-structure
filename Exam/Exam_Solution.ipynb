```python
from CPE112_Module import Array2D
# Shape Instant Variables -> Rows , Cols 
class Test(Array2D):
    def __init__(self,numRows,numCols):
        super().__init__(numRows,numCols)
        self._Shape = numRows, numCols
        self._sum = numRows + numCols
    
    def load(self,a_list):
        assert (len(a_list) ==  self.numRows()) and ( len(a_list[0]) == self.numCols()) , "Error size doesn't clear"
        
        for i in range(self.numRows()):
            for j in range(self.numCols()):
                self[i,j] = a_list[i][j]
        print(self)

    # 2 x 3 -> 3 x 2
    # [ [1,2,3] , [4,5,6]] -> [ [1,2] , [3,4] , [5,6]]

    def reshape(self,new_Rows,new_Cols):
        assert (self.numRows() * self.numCols() == new_Rows * new_Cols) , "Error not compatible size."
        
        # Step 1 
        number_list = []
        ndx = 0
        for i in range(self.numRows()):
            for j in range(self.numCols()):
                number_list.append(self[i,j])
        print(number_list)

        newMatrix = Test(new_Rows,new_Cols)
        for i in range(new_Rows):
            for j in range(new_Cols):
                newMatrix[i,j] = number_list[ndx]
                ndx += 1

        return newMatrix
```

```python
from CPE112_Module import Matrix

class Test2(Matrix):
    def trace(self):
        total = 0
        for i in range(self.numRows()):
            total += self[i,self.numRows()-1-i]
        return total
    
    def det(self):
        assert (self.numRows() == 2 and self.numCols() == 2 ) or (self.numRows() == 3 and self.numCols() == 3) , "Error, not a square matrix."
        # 2x2
        if (self.numRows() == 2 and self.numCols() == 2):
            total = self[0,0] * self[1,1] - self[0,1] * self[1,0]
            return total
        
        # 3x3
        elif (self.numRows() == 3 and self.numCols() == 3):
            sum_total = 0 
            # บวก คูณลง
            for i in range(self.numRows()):
                sum_total += self[i%3,i%3] * self[(i+1)%3,(i+1)%3] * self[(i+2)%3,(i+2)%3]
            
            # ลบ คูณขึ้น
            min_total = 0
            for i in range(self.numRows()):
                min_total += self[i%3,self.numRows()-1-i] * self[(i+1)%3,self.numRows()-i] * self[(i+2)%3,self.numRows()-i+1]

            return sum_total - min_total
        
    def inverse(self):
        assert (self.numRows() == 2 and self.numCols() == 2) , "Only receive 2x2 matrix."
        scalar = 1 / (self[0,0] * self[1,1] - self[0,1] * self[1,0])
        newMatrix = Test2(2,2)
        newMatrix[0,0] = self[1,1]
        newMatrix[0,1] = - self[0,1]
        newMatrix[1,0] = - self[1,0]
        newMatrix[1,1] = self[0,0]
        newMatrix.scaleBy(scalar)
        return newMatrix
    
    def submatrix(self,row,col):

        new_list = []
        # row , col - ลบแถว หลักทิ้ง
        for i in range(self.numRows()):
            for j in range(self.numCols()):
                if (i == row-1) or (j == col-1):
                    continue
                else:
                    new_list.append(self[i,j])
        
        newMatrix = Test2(row-1,col-1)
        ndx = 0
        # เข้าไป Assign ค่า
        for i in range(row-1):
            for j in range(col-1):
                newMatrix[i,j] = new_list[ndx]
                ndx += 1
        return newMatrix 
```
