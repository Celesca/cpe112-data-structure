## Line Plots

Plot serise of data from Python's list or Numpy's array

```python
x = np.linspace(0,15,30) #linspace(start, stop, Number of samples to generate)
y = np.sin(x) + 0.1*np.random.randn(len(x))#sin(x) + 0.1(random_value)
```

```python
plt.plot(x,y)
```

Plot graph with custom options <br>
e.g.; plot with dash line and circle, purple color, line width = 2, marker side = 10

```python
plt.plot(x,y,color='blue', linestyle='-', linewidth='3', marker='o',markersize =7)
```

```python
plt.plot(x,y,'o--', color='purple', lw=2, ms=10)
```

Set Figure size

```python
plt.figure(figsize=(8,3), dpi=60)#figure(figsize=(แนวนอน, แนวตั้ง), dpi=int) 
plt.plot(x,y)
```

Add Axis labels

```python
plt.figure(figsize=(8,3))
plt.plot(x,y)
plt.xlabel('Time [s]', fontsize=16)
plt.ylabel('Voltage [V]')
plt.title('Voltage in 3rd Electrode')
plt.show()
```

Add Legend

legend? 
คือ คำอธิบายสัญลักษณ์หรือสีของชุดข้อมูลในกราฟ

```python
plt.figure(figsize=(8,3))
plt.plot(x,y, label='Component 1')
plt.xlabel('Time [s]', fontsize=16)
plt.ylabel('Voltage [V]')
plt.title('Voltage in 3rd Electrode')
plt.legend(loc='lower right', fontsize=18)
plt.show()
```

## Histograms

```python
res = np.random.randn(1000)*0.2 + 0.4
```

```python
plt.figure(figsize=(8,3))
plt.hist(res)
plt.show()
```

Change number of bins

```python
plt.figure(figsize=(8,3))
plt.hist(res, bins=30)
plt.show()
```

Convert to density plot

```python
plt.figure(figsize=(8,3))
plt.hist(res, bins=30, density=True)
plt.show()
```

Differnt style of bins (useful for plotting multiple histograms)

```python
res2 = np.random.randn(1000)*0.2 + 0.4
```

```python
plt.figure(figsize=(8,3))
plt.hist(res, bins=30, density=True, histtype='step')
plt.hist(res2, bins=30, density=True, histtype='step')
plt.xlabel('$\hat{E}-E$', fontsize=20)
plt.ylabel('Frequency')
plt.show()
```

## Multiple subplots on 1 figure


## Single Axes in one Figure

```python
fig, ax = plt.subplots(1, 1, figsize=(12,4))
ax.plot(x, y, 'o--', color='r', lw=0.4, ms=3)
ax.text(0.1, 0.1, 'text here', transform=ax.transAxes)
ax.set_xlabel('Time [s]')
ax.set_ylabel(r'$\frac{d}{dx} f(x)$', fontsize=15)
plt.show()
```

## Multiple axes in 1 figure

```python
fig, axes = plt.subplots(3, 2, figsize=(8,8))
ax = axes[0][0]
ax.plot(x, y, 'o--', color='r', lw=0.4, ms=3)
ax.text(0.1, 0.1, 'text here', transform=ax.transAxes)
ax = axes[1][1]
ax.hist(res, bins=30, density=True, histtype='step', label='C1')
ax.hist(res2, bins=30, density=True, histtype='step', label='C1')
ax.set_xlabel('$\hat{E}-E$')
ax.legend()
fig.tight_layout()
fig.suptitle('Title of All Plots', y=1.03, fontsize=25)
plt.show()
```

Display x-ticks

```python
#plt.style.use(['science', 'notebook', 'grid'])
fig, ax = plt.subplots(1, 1, figsize=(8,3))
ax.plot(x, y, 'o-', color='r', lw=1, ms=5)
ax.text(0.1, 0.1, 'text here', transform=ax.transAxes)
ax.set_xlabel('Time [s]')
ax.set_ylabel(r'$\frac{d}{dx} f(x)$', fontsize=15)
ax.tick_params(axis='both', which='major', labelsize=10)
ax.grid('minor')
plt.show()
```

# Example for Civil Engineering 

## Line Plot

```python
# plotting stress-strain curves.
import matplotlib.pyplot as plt
import numpy as np

# 1.Generate sample data for stress-strain curve
strain = np.linspace(0, 0.1, 100)  # Strain values from 0 to 0.1
stress = 200 * strain  # Simple linear relationship (for illustration purposes)

# 2.Plot stress-strain curve
plt.figure(figsize=(8, 6))
plt.plot(strain, stress, label='Stress-Strain Curve', color='blue', linewidth=2)

# 3.Add labels and title
plt.title('Stress-Strain Curve for a Material')
plt.xlabel('Strain')
plt.ylabel('Stress (MPa)')

# 4.Add grid
plt.grid(True, linestyle='--', alpha=0.7)

# 5.Add legend
plt.legend()

# 6.Show the plot
plt.show()
```

## 3D Plot

```python
# 3D Plot for Site Topography:
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Generate sample topography data
x = np.linspace(0, 10, 100)
y = np.linspace(0, 10, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Plotting a 3D surface plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('Site Topography')
plt.show()
```
